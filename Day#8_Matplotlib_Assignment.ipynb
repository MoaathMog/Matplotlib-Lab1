## Q2: Create pie chart with equal explode values.

```python
movies = ['Avatar', 'Star Wars', 'Titanic', 'The Avengers', 'Jurassic World']
rating = [70,88,89,92,89]

plt.pie(rating ,labels=movies,autopct='%1.1f%%', startangle=90, radius=1, explode=[0,0,0.1,0,0.1])
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
np.random.seed(2222)
num =20
x = np.random.rand(num)
y = np.random.rand(num)
colors = np.random.rand(num)
plt.scatter(x, y, c=colors, alpha=0.4, s=50)
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(x, y, c=colors, alpha=0.4, s=50)
plt.title('Random dots',c='red',fontsize = 20,loc='left')
plt.xlabel('Distance',fontsize = 16)
plt.ylabel('Hight',fontsize = 16)
```
